```python
import numpy as np
import matplotlib.pyplot as plt

# Define the neuron integration function
def ode(u,v):
            u_i=u
            v_i=v
            dxdt = [10*((1)*(u_i*(u_i-0.4)*(1-u_i))-v_i   ),
                    
                    ((u_i*0.04-0.16*v_i))
                    
                    ]
            return dxdt


# Fixed pairs of initial conditions for u and v
u_initials = [0.5]
v_initials = [0]
iapp_values = [ 1]
fig, ax = plt.subplots(1, 2, figsize=(10, 6))

# Function to solve the system
def solve_system(DT, plot_interval,i=0):
    num_steps = int(20 / DT)
    time_points = np.arange(0, 20, DT)

   
    for u_init, v_init, iapp in zip(u_initials, v_initials, iapp_values):
        u = np.zeros(num_steps)
        v = np.zeros(num_steps)
        u[0] = 0.5
        v[0] = 0
        
        for step in range(1, num_steps):
            du,dv= ode(u[step-1], v[step-1])
            u[step]=du*DT+u[step-1]
            v[step]=dv*DT+v[step-1]
        # Plot every 'plot_interval' time steps
        ax[i].plot(time_points[::plot_interval],u[::plot_interval],"o", label="$U$")
    
        ax[i].set_xlabel('Time')
        if(i==0):
           ax[i].set_ylabel('Scarce training data set')
        else:
            ax[i].set_ylabel('Large training data set')
        ax[i].legend()
    
    
    fig.show()
    return u[::plot_interval]

# Solve and plot for DT = 0.01, plotting every 1 unit of time
u_t=solve_system(DT=0.01, plot_interval=100,i=0)

# Solve and plot for DT = 0.1, plotting every 1 unit of time
u_p=solve_system(DT=0.1, plot_interval=10,i=1)

print(np.mean(np.sum((u_t-u_p)**2)**(0.5)))
plt.savefig("scarce-plenty.pdf")
```

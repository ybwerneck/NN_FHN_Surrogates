```python
import matplotlib.pyplot as plt

def plot_model_errors_vs_neurons(df, error_col, neuron_col, shape_col, pinn_col):
    """
    Plots model errors against total neurons with customization for shape and color.

    Parameters:
        df (pd.DataFrame): DataFrame containing the data.
        error_col (str): Column name for error values.
        neuron_col (str): Column name for total neurons.
        shape_col (str): Column name for model shapes.
        pinn_col (str): Column name indicating PINN info (True/False).

    Returns:
        None
    """
    # Define color mapping for PINN Info
    colors = df[pinn_col].map({True: 'blue', False: 'orange'})

    # Define marker shapes based on model shape
    shape_markers = {'funnel': 'o', 'rectangle': 's', 'bottleneck': 'D'}

    # Create the plot
    plt.figure(figsize=(10, 6))
    for shape in df[shape_col].unique():
        shape_mask = (df[shape_col] == shape)
        plt.scatter(
            df[neuron_col][shape_mask], df[error_col][shape_mask],
            s=25,  # Size of the dots
            c=colors[shape_mask],
            label=f'Shape: {shape}',
            marker=shape_markers.get(shape, 'o')  # Default marker is 'o'
        )

    # Customize the plot
    plt.title('Model Errors vs. Total Neurons', fontsize=14)
    plt.xlabel('Total Neurons', fontsize=12)
    plt.ylabel('Final Error 1', fontsize=12)
    plt.yscale("log")
    plt.legend(title="Model Shape", fontsize=10)
    plt.grid(alpha=0.3)
    plt.show()

# Example Usage:
# Assuming `df` is your DataFrame
plot_model_errors_vs_neurons(models, error_col="Final Error 1", neuron_col="total_neurons", shape_col="shape", pinn_col="Pinn Info")



```
